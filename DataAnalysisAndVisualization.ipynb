```python
import pandas as pd

### 1- loading data
df = pd.read_csv('imdb_top_1000.csv')

### 2- Clean and preprocess the data by :
## 1) Remove unnecessary columns
# we will only need: Series_Title,Genre,Director
# so we can drop: Poster_Link,Released_Year,Certificate,Runtime,IMDB_Rating,Overview,Meta_score,Star1,Star2,Star3,Star4,No_of_Votes,Gross
df = df.drop(['Poster_Link', 'Released_Year', 'Certificate', 'Runtime','IMDB_Rating', 'Overview', 'Meta_score', 'Star1', 'Star2', 'Star3', 'Star4', 'No_of_Votes', 'Gross'], axis=1)

## 2) Removing the missing values
df = df.dropna()

### 3- The 'Genre' column must be separated to multiple genres
df['Genre'] = df['Genre'].str.split(',')

# Make a new dataframe that store the individual (separated) genres with their counts
genre_counts = pd.DataFrame(df['Genre'].explode().str.strip().value_counts()).reset_index()
# Rename the columns
genre_counts.columns = ['Genre', 'Count']

### 4- Identify the most popular genre 
popular_genre = genre_counts.loc[genre_counts['Count'].idxmax(), 'Genre']
popular_genre_count = genre_counts.loc[genre_counts['Count'].idxmax(), 'Count']
print(f"The most popular genre is {popular_genre} with a count of {popular_genre_count}.")


### 5- Visualize the data using a bar chart to show the most popular genre
import matplotlib.pyplot as plt
plt.figure(figsize=(40, 7))
plt.bar(genre_counts['Genre'], genre_counts['Count'])
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('The Most Popular Genre in Movies/TV views')
plt.show()


### 6- Identify the director with the most top-rated movies/TV shows
director_counts = df['Director'].value_counts().reset_index()
director_counts.columns = ['Director', 'Count']
top_director = director_counts.loc[director_counts['Count'].idxmax(), 'Director']
print('The director with the most top-rated movies/TV shows: ',top_director)

### 7- Visualize the data that show the director...
plt.figure(figsize=(100, 6))
plt.bar(director_counts['Director'], director_counts['Count'])
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Director with the Most Top-rated Movies/TV Shows')
plt.xticks(rotation=90)
plt.show()
```

## Analysis Report: Top Movies and TV Shows

### Most Popular Genre

To identify the most popular genre of movies and TV shows, we analyzed the Genre column. As some entries had multiple genres separated by commas, we split the column to separate the genres. We then used the value_counts() function to count the occurrences of each genre. The genre with the highest count was determined to be the most popular genre.

### Insight

The most popular genre among the top movies and TV shows in the dataset is Drama.

### Director with the Most Top-rated Movies/TV Shows

We further analyzed the Director column to identify the director with the most top-rated movies/TV shows. We counted the occurrences of each director and found the director with the highest count.

### Insight
